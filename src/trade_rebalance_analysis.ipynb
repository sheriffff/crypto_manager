```python
import numpy as np
import matplotlib.pyplot as plt

from src.trade_rebalance import analyze_and_trade, get_target_proportion_dollars_eth
```

Rebalancing strategy (depends on params set in `src/trade_rebalance.py`):
 - I always hold at least 20% of my portfolio in ETH, even if the price is very high (above $5000).

 - I always have at least 10% liquid cash, even if the price is very low (below $3000).
 - As price increases, the target proportion of ETH decreases, meaning I should sell some ETH to maintain balance.

```python
prices = np.linspace(3000, 5000, 100)
target_proportions = [get_target_proportion_dollars_eth(price) for price in prices]

plt.figure(figsize=(10, 6))
plt.plot(prices, target_proportions, label='Target ETH Proportion', color='red')
plt.title('ETH Proportion vs Price')
plt.xlabel('ETH Price ($)')
plt.ylabel('ETH Proportion')
plt.ylim(0, 1)
plt.legend()
plt.grid()
plt.show()
```
